```python
from IPython.display import display, Markdown, clear_output
# widget packages
import ipywidgets as widgets
```

```python
# Widgety do wpisania koordynatów wierzchołków obszaru

coords_wall1_x = widgets.BoundedIntText(
    value=0,
    description='Type x1:',
    disabled=False
)
coords_wall1_y = widgets.BoundedIntText(
    value=0,
    description='Type y1:',
    disabled=False
)
coords_wall2_x = widgets.BoundedIntText(
    value=3,
    description='Type x2:',
    disabled=False
)
coords_wall2_y = widgets.BoundedIntText(
    value=0,
    description='Type y2:',
    disabled=False
)
coords_wall3_x = widgets.BoundedIntText(
    value=3,
    description='Type x3:',
    disabled=False
)
coords_wall3_y = widgets.BoundedIntText(
    value=3,
    description='Type y3:',
    disabled=False
)
coords_wall4_x = widgets.BoundedIntText(
    value=0,
    description='Type x4:',
    disabled=False
)
coords_wall4_y = widgets.BoundedIntText(
    value=3,
    description='Type y4:',
    disabled=False
)

mywall = []  # inicjujemy tablicę koordynatów obszaru

wall_button = widgets.Button(description='Confirm') # tworzymy przycisk i miejsce na wyświetlenie outputu
wall_out = widgets.Output()

# Tworzymy event on_click

def on_button_clicked(_):
    # Gdy naciśniemy przycisk wall_button:
      with wall_out:
          # Czyścimy output, aktualizujemy koordynaty wierzchołków obszaru i wyświetlamy nowy output
          clear_output()
          global mywall
          mywall= np.array([[int(coords_wall1_x.value),int(coords_wall1_y.value)],[int(coords_wall2_x.value),int(coords_wall2_y.value)],[int(coords_wall3_x.value),int(coords_wall3_y.value)],[int(coords_wall4_x.value),int(coords_wall4_y.value)]],dtype=object)
          print('Wall settings changed.')
          print(mywall)

wall_button.on_click(on_button_clicked) # Przypisujemy event
```

```python
# Widgety do wpisania koordynatów wierzchołków przeszkód

coords_obstacle1_x = widgets.BoundedIntText(
    value=1,
    description='Type x1:',
    disabled=False
)
coords_obstacle1_y = widgets.BoundedIntText(
    value=1,
    description='Type y1:',
    disabled=False
)
coords_obstacle2_x = widgets.BoundedIntText(
    value=2,
    description='Type x2:',
    disabled=False
)
coords_obstacle2_y = widgets.BoundedIntText(
    value=1,
    description='Type y2:',
    disabled=False
)
coords_obstacle3_x = widgets.BoundedIntText(
    value=2,
    description='Type x3:',
    disabled=False
)
coords_obstacle3_y = widgets.BoundedIntText(
    value=2,
    description='Type y3:',
    disabled=False
)
coords_obstacle4_x = widgets.BoundedIntText(
    value=1,
    description='Type x4:',
    disabled=False
)
coords_obstacle4_y = widgets.BoundedIntText(
    value=2,
    description='Type y4:',
    disabled=False
)

myobstacles = np.array([]) # inicjujemy tablicę koordynatów przeszkód

obstacle_button = widgets.Button(description='Confirm') # tworzymy przycisk zatwierdzania nowej przeszkody

def on_button_clicked(_):
    # Gdy naciśniemy przycisk obstacle_button ('Confirm'):
      with output:
          clear_output() # Czyścimy output
          global myobstacles
          myobstacles = list(myobstacles) # Zmieniamy typ tablicy z przeszkodami na standardową listę Pythona
          myobstacles.append([[int(coords_obstacle1_x.value),int(coords_obstacle1_y.value)],[int(coords_obstacle2_x.value),int(coords_obstacle2_y.value)],[int(coords_obstacle3_x.value),int(coords_obstacle3_y.value)],[int(coords_obstacle4_x.value),int(coords_obstacle4_y.value)]]) # Dodajemy kolejną przeszkodę
          print('Obstacle added.')  # Wyświetlamy output
          print(myobstacles)
            
removeobstacle_button = widgets.Button(description='Remove last obstacle') # tworzymy przycisk usuwania ostatniej przeszkody
            
def on_removeobstacle_clicked(_):
    # Gdy naciśniemy przycisk removeobstacle_button ('Remove last obstacle'):
      with output:
          clear_output() # Czyścimy output
          global myobstacles
          myobstacles = list(myobstacles) # Zmieniamy typ tablicy z przeszkodami na standardową listę Pythona
          myobstacles = myobstacles[:-1] # Usuwamy ostatni element tablicy
          print('Obstacle removed.') # Wyświetlamy output
          print(myobstacles)

obstacle_button.on_click(on_button_clicked) # Przypisujemy eventy
removeobstacle_button.on_click(on_removeobstacle_clicked)
```
